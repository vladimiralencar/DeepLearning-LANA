# Pesquisando uma imagem dentro de outra imagem

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

# Carrega a iaagem e converte para Grayscale
image = cv2.imread('images/WaldoBeach.jpg')

# correct BGR to RGB
imagecolor = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# show image
plt.figure(figsize=(12,8))
plt.title('Onde est Waldo?')
plt.imshow(imagecolor)

# Carrega a imagem que queremos buscar na outra imagem
template = cv2.imread('images/waldo.jpg',0)
template2 = cv2.imread('images/waldo.jpg')
# correct BGR to RGB
templatecolor = cv2.cvtColor(template2, cv2.COLOR_BGR2RGB)
plt.figure()
plt.title('Waldo')
plt.imshow(templatecolor)

# converte a imagem principal para escala de cinza
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Faz o match
result = cv2.matchTemplate(gray, template, cv2.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

# Cria a Bounding Box
top_left = max_loc
bottom_right = (top_left[0] + 50, top_left[1] + 50)
                                             # cor     #largura
cv2.rectangle(image, top_left, bottom_right, (0,0,255), 5)

# correct BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# show image
plt.figure(figsize=(12,8))
plt.title('Onde est Waldo?')
plt.imshow(image)
```
